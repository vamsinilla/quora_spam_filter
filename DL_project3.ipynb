```
train = pd.read_csv(r'/content/train.csv')
```

```
del train['qid']
```

```
from sklearn.model_selection import train_test_split
```

```
spam_train,spam_test = train_test_split(train,test_size =0.2,random_state =2)
```

```
from numpy import array
from keras.preprocessing.text import one_hot
from keras.preprocessing.sequence import pad_sequences
from keras.models import Sequential
from keras.layers.core import Activation, Dropout, Dense
from keras.layers import Flatten, LSTM
from keras.layers import GlobalMaxPooling1D
from keras.models import Model
from keras.layers.embeddings import Embedding
from sklearn.model_selection import train_test_split
from keras.preprocessing.text import Tokenizer
from keras.layers import Input
from keras.layers.merge import Concatenate
```

```
x_train = spam_train['question_text']
x_test = spam_test['question_text']
y_train = spam_train['target']
y_test = spam_test['target']
```

```
import nltk
nltk.download('punkt')
```

```
from nltk import word_tokenize
```

```
sent_lens=[]

for sent in x_train:
  sent_lens.append(len(word_tokenize(sent)))
```

```
max(sent_lens)
```

```
np.quantile(sent_lens,0.9)
```

```
max_len = 30
```

```
tok = Tokenizer(char_level=False,split=' ')
tok.fit_on_texts(x_train)
sequences_train = tok.texts_to_sequences(x_train)
```

```
sequences_train
```

```
len(tok.index_word)
```

```
vocab_len = len(tok.index_word.keys())
```

```
vocab_len
```

```
sequences_matrix_train = pad_sequences(sequences_train, maxlen=max_len)
```

```
sequences_test = tok.texts_to_sequences(x_test)
```

```
sequences_matrix_test = pad_sequences(sequences_test, maxlen=max_len)
```

```
def RNN():
  inputs = Input(name = 'inputs',shape=[max_len])
  layer = Embedding(vocab_len+1,500,input_length=max_len,mask_zero=True)(inputs)
  layer = LSTM(64)(layer)
  layer = Dense(256,name= 'FC1')(layer)
  layer = Activation('relu')(layer)
  layer = Dropout(0.5)(layer)
  layer = Dense(1,name = 'output_layer')(layer)
  layer = Activation('sigmoid')(layer)
  model = Model(inputs=inputs,outputs= layer )
  return model
```

```
model = RNN()
```

```
model.summary()
```

```
model.compile(loss= 'binary_crossentropy',optimizer='adam',metrics=['accuracy'])
```

```
from tensorflow.keras.callbacks import ModelCheckpoint
import os
```

```
output_folder = './model_output'

if not os.path.exists(output_folder):
  os.makedirs(output_folder)
```

```
filepath = output_folder+"/weights-{epoch:02d}-{accuracy:.4f}.h5"
checkpoint = ModelCheckpoint(filepath,monitor='accuracy',verbose=1,save_best_only=False,save_weights_only=True,mode='auto',period = 3)
```

```
from tensorflow.keras.callbacks import EarlyStopping

earlystop = EarlyStopping(monitor= 'loss',min_delta= 0.0001,patience =4,verbose=2,mode='auto')
```

```
model.fit(sequences_matrix_train,y_train,batch_size= 5000,validation_data=(sequences_matrix_test,y_test),callbacks=[checkpoint])
```

```
predictions = model.predict(sequences_matrix_test)
from sklearn.metrics import roc_auc_score
roc_auc_score(y_test,predictions)
```
